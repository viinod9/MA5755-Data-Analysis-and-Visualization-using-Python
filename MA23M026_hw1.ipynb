Count the number of vowels in a given string.

```python
stri = input("Enter string : ")
stri1 = stri.lower()
count = 0
for i in stri1:
  if (i== 'a'or i == 'e'or i == 'i'or i == 'o'or i == 'u'):
    count+=1
print(count)
```
