###(1) : Write a program to print numbers divisible by both 7 and 8. Numbers between 1 and 1000.

Numbers like 56, which are both divisible by 7 and 8
###(a): Using continue statement

```python
result = []
for i in range(1,1001):
  div_8 = i%8
  div_7 = i%7
  if(div_8 == 0 and div_7 == 0):
    result.append(i)
  else:
    continue
print("Number divisible by 7 and 8 , using continue are : \n",result)
```

###(2): Number 34567222541147 has been given. Print the addition of all even digits and the addition of all odd digits.

(Sample input = 156893, Sample Output: sum_even = 14 , sum_odd = 18)

```python
num = 34567222541147
even_num = 0
odd_num = 0
while (num!=0):
  remain = num%10
  if(remain%2 == 0):
    even_num += remain
  else:
    odd_num += remain
  num = num//10

print("Sum of odd digits : ",odd_num)
print("Sum of even digits : ",even_num)
```

### (3): Describe conditional expression, Write a program to print the maximum and minimum numbers from a list using conditional expression.

 Use the max () and min () functions to verify your output.

```python
lis = [10 , 34 , 67 , 23 ,89 , 37 , 67 , 396]
max1 = lis[0]
min1 = lis[0]
for num in lis:
  if (num > max1):
    max1 = num
  if (num < min1):
    min1 = num

print("Max of above list is : ",max1)
print("Min of above list is : ",min1)
print("Results after using max , min functions :", max(lis) , min(lis))
```

###(5) A palindrome sentence is a phrase or sentence that reads the same forward as it does backward, ignoring spaces, punctuation, and capitalization.

(A) Take a number as input, determine if it is a palindrome, considering only integers. Eg: 2442

```python
num = int(input("Enter the number you want to check for palandrome :"))
n = num
rev = 0
while (n>0):
  remain = n%10
  rev = (rev*10) + remain
  n = n//10
if (num == rev):
  print('Palandrome')
else:
  print('Not Palandrome')

```

(B) Now, take a string as an input and check whether it is Palindrome or not.  

eg: "Was it a car or a cat I saw"

The expected output for this example will print, 'True'

```python
string = "Was it a car or a cat I saw?"
string1 = ''.join(c.lower() for c in string if c.isalnum())
if (string1 == string1[::-1]):
  print("True")
else:
  print("False")
```

### 7.You have a single list of names (each name should have first name and last name). Write a programme

(a) to print the first names.

(b) to print the last names.

(c) to print each name as 'last_name first_name.

(d) to print the names in the sorted order with respect to last name

Example :Input list: ['Mark Antony', 'Mohan Raj'],

Output :

 (a) ['Mark', 'Mohan']  

(b) ['Antony', 'Raj']

(c) ['Antony Mark', 'Raj Mohan']

(d) ['Mark Antony', 'Mohan Raj']




```python
input_list = ['Mark Antony', 'Mohan Raj']

first_names = []
last_names = []
for name in input_list:
    first_names.append(name.split()[0])
    last_names.append(name.split()[1])
print("First names:", first_names)
print("Last names:", last_names)

formatted_names = []
for name in input_list:
    formatted_names.append(' '.join(name.split()[::-1]))
print("Formatted names:", formatted_names)

sorted_names1 = sorted(formatted_names)
sorted_names = []
for ele in sorted_names1:
  name = ele.split()
  sorted_names.append(name[1]+' '+name[0])
print("Sorted names:", sorted_names)
```

### (8). WAP to find the minimum of 3 numbers using
