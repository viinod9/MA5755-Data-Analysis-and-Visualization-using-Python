```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Define the data points
data = np.array([
    [0, 0, 'A'], [0, 1, 'A'], [0, 2, 'A'], [0, 3, 'A'], [0, 4, 'A'],
    [1, 0, 'A'], [1, 1, 'B'], [1, 2, 'B'], [1, 3, 'B'], [1, 4, 'A'],
    [2, 0, 'A'], [2, 1, 'B'], [2, 2, 'A'], [2, 3, 'B'], [2, 4, 'A'],
    [3, 0, 'A'], [3, 1, 'B'], [3, 2, 'B'], [3, 3, 'B'], [3, 4, 'A'],
    [4, 0, 'A'], [4, 1, 'A'], [4, 2, 'A'], [4, 3, 'A'], [4, 4, 'A']
])

# Map class labels to numerical values
label_map = {'A': 0, 'B': 1}
data[:, 2] = [label_map[label] for label in data[:, 2]]

# Separate features and labels
X = data[:, :2].astype(int)
y = data[:, 2].astype(int)

# Fit a KNN classifier with K=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)

# Plot the data points
plt.figure(figsize=(6, 6))
for i, label in enumerate(np.unique(y)):
    X_class = X[y == label]
    plt.scatter(X_class[:, 0], X_class[:, 1], label=label)

# Plot decision boundaries
h = .02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower', alpha=0.2)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('KNN Decision Boundaries (K=1)')
plt.legend()
plt.grid(True)
plt.show()
```
